```python
'''
Générateur de séquence aléatoires utilisant les processeurs quantiques d'IBM

1ère case: setup (regarder les disponibilités pour choisir l'appareil réel)

Après le set-up, éxécuter la deuxième case pour obenir et ajouter 375 bits aléatoires au .txt

'''
from qiskit import *

#Ne fonctionne que si un compte IBM a déjà été enregistré
IBMQ.load_account()


q = QuantumRegister(5)
c = ClassicalRegister(5)

circuit = QuantumCircuit(q, c)
for i in range(5):
    circuit.h(i)
    circuit.measure(i,i)

provider = IBMQ.get_provider(hub='ibm-q')
for backend in provider.backends():
    print( backend.status() )

#    print("\n")
#    print( backend.configuration() )
#    print("______________\n")

    
'''
Tous les réels: pas plus de 75 experiences par execution
'''
```

```python
real_device = provider.get_backend('ibmq_london')

qc_sequence = ""
circuits=[]
nb_circ=75
for i in range(nb_circ):
    circuits.append(circuit)

    
job = execute(circuits, backend = real_device, shots=1)

for i in range(nb_circ):
    result = job.result().get_counts(i)
    for resultat in result:
        qc_sequence += resultat
    
fichier = open("sequence_qc.txt", "a")
fichier.write(qc_sequence)
fichier.close()
```

```python
"""
Génération de séquence grâce à ANU Quantum Random Numbers Server

"""
import requests,json

#max pour les 2: 1024
taille_string = 30
nb_string = 1024

#La requête nous renvoie nb_string de string dans une liste, chacun ayant taille_string nombres en hexa
r = requests.get("https://qrng.anu.edu.au/API/jsonI.php?length="+str(nb_string)+"&type=hex16&size="+str(taille_string))
donnees = json.loads(r.text)

#On convertit le résultat en bits dans un string et l'enregistre dans un fichier.
string_final = ""
for i in range(nb_string):
    string = donnees["data"][i]
    for j in range(taille_string):
        nb = int("0x"+string[2*j]+string[2*j+1],16)
        string_final += bin(nb)[2:]

fichier = open("sequence_QRGN_ANU.txt", "w")
fichier.write(string_final)
fichier.close()
```

```python
"""
Formatage d'une séquence obtenue grâce à Nano-optics groups du départment de physique de l'université de Humboldt

Téléchargée sur cette page: https://qrng.physik.hu-berlin.de/download (besoin d'être identifié)

La séquence téléchargée est en .bin, ce bloc est la pour le convertir en séquence binaire sur un .txt
"""

fichier = open("sampledata-1MB.bin", "rb")
string = fichier.read()
fichier.close()


string_final = ""
for char in string:
    string_final += bin(char)[2:]

fichier = open("sampleDataBerlin.txt", "w")
fichier.write(string_final)
fichier.close()
```
